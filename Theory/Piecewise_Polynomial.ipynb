### Changepoint Detection on polynomial functions

- We wish to determine a changepoint in polynomial functions, beginning with the simple case of constant and linear
- Ideally what if we developed a breakpoint detection algorithm which searched for breakpoints across a series of basis expansions and then applied BIC/AIC to determine the best fit, with a penalizer on complexity of terms
- Segmentation and a fitting problem



### HOPS
- One approach that supports more complex polynomial terms is HOPS -> High ordered polynomial segmentation which utilizes fits and consequently segments a signal into polynomials of k'th order
- MATLAB codes https://www.mathworks.com/matlabcentral/fileexchange/98504-hops-a-fast-algorithm-for-segmenting-piecewise-polynomials

### Mixing Terms 
- One way we can approach this is by solving an OLS system for best fitted polynomial on each segmentation
- Utilizing a penalized least squares we can approximate the best fitting polynomial which might be different ordered -> does shrinkage to correlation correspond to different ordered estimates
